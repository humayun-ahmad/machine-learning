Dropout Regularization In Deep Neural Network
This is a dataset that describes sonar chirp returns bouncing off different services. The 60 input variables are the strength of the returns at different angles. It is a binary classification problem that requires a model to differentiate rocks from metal cylinders.

Dataset information: https://archive.ics.uci.edu/ml/datasets/Connectionist+Bench+(Sonar,+Mines+vs.+Rocks) Download it from here: https://archive.ics.uci.edu/ml/machine-learning-databases/undocumented/connectionist-bench/sonar/sonar.all-data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("./sonar_dataset.csv", header=None)
df.sample(5)
```

```python
df.shape
```

```python
df.isnull().sum()
```

Split the data

```python
X = df.drop(60, axis='columns')
y = df[60]

X
```

```python
y
```

```python
y = pd.get_dummies(y, drop_first=True)
y.sample(5)
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(60, activation='relu', input_shape=(60,)),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')  # sigmoid for binary classification
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=100, batch_size=8)
```

```python
model.evaluate(X_test, y_test)
```

```python
# y_test.shape.reshape(-1)

y_pred = model.predict(X_test)

arr = np.array(y_pred)
print(arr.shape)


y_pred = model.predict(X_test).reshape(-1)

arr = np.array(y_pred)
print(arr.shape)
```

```python
y_pred = model.predict(X_test).reshape(-1)
print(y_pred[:10])

y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(y_test, y_pred))
```

Now try to use Dropout Layer

```python
model = keras.Sequential([
    keras.layers.Dense(60, activation='relu', input_shape=(60,)),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1, activation='sigmoid')  # sigmoid for binary classification
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, epochs=100 , batch_size=8)
```

```python
model.evaluate(X_test, y_test)
```

```python
y_pred = model.predict(X_test).reshape(-1)
print(y_pred[:10])

y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(y_test, y_pred))
```
