# Probability Distribution

Normal Distribution

```python
rank_3 = tf.Variable([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])

print(rank_3.shape)

rank_4 = tf.reshape(tf.range(start=0, limit=24, delta=1), shape=[2, 3, 2, 2])
print(rank_4)
```

```python
from matplotlib import pyplot as plt
N = 100
normal_dist = tf.random.normal(shape=(N,), mean=0, stddev=2.)
plt.plot(normal_dist)
plt.xlabel('Random number idx')
plt.ylabel('N random value')
plt.show()
```

Uniform Distribution

```python
uniform_dist = tf.random.uniform(shape=(4,4), minval=0, maxval=10, dtype='int32')
print(uniform_dist)
plt.plot(uniform_dist)
plt.xlabel('Random number idx')
plt.ylabel('N Random number')
plt.show()
```

## GradientTape

```python
import tensorflow as tf

x = tf.constant(2.0)
print(x)
with tf.GradientTape() as tape:
    tape.watch(x)
    y = x**2 + 2*x + 5
    
# `x` এর সাপেক্ষে `y` এর গ্র্যাডিয়েন্ট কত ?
gradient = tape.gradient(y,x)
```
